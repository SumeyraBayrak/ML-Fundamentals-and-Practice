```python
def format_column_names(df):
    """
    Format column names to be lowercase with underscores between words.
    Removes special characters and converts spaces to underscores.
    
    Parameters:
    df (pandas.DataFrame): Input DataFrame with column names to format
    
    Returns:
    pandas.DataFrame: DataFrame with formatted column names
    
    Example:
    >>> df = pd.DataFrame({'Column Name': [1], 'Another.Column!': [2]})
    >>> format_column_names(df)
    # Returns DataFrame with columns: 'column_name', 'another_column'
    """
    def clean_column_name(col):
        # Convert to lowercase
        col = col.lower()
        
        # Replace special characters and multiple spaces with a single space
        col = re.sub(r'[^a-z0-9\s]', ' ', col)
        col = re.sub(r'\s+', ' ', col)
        
        # Strip leading/trailing spaces and replace remaining spaces with underscores
        col = col.strip().replace(' ', '_')
        
        # Remove consecutive underscores
        col = re.sub(r'_+', '_', col)
        
        return col
    
    # Create a copy of the DataFrame to avoid modifying the original
    df_formatted = df.copy()
    
    # Apply the formatting to all column names
    df_formatted.columns = [clean_column_name(col) for col in df_formatted.columns]
    
    return df_formatted
```

```python
def clean_categorical_values(df, categorical_columns):
    """
    Clean categorical values by:
    1. Converting to lowercase
    2. Replacing special characters with spaces
    3. Removing extra spaces
    4. Converting remaining spaces to underscores
    5. Removing consecutive underscores
    
    Args:
        df: pandas DataFrame
        categorical_columns: list of column names to clean
    
    Returns:
        DataFrame with cleaned values
    """
    
    df_clean = df.copy()
    
    for column in categorical_columns:
        # Convert to string type (in case of numeric values)
        df_clean[column] = df_clean[column].astype(str)
        
        # Apply cleaning steps
        df_clean[column] = (df_clean[column]
            .str.lower()  # Convert to lowercase
            .str.replace(r'[^a-z0-9\s]', ' ', regex=True)  # Replace special chars with space
            .str.replace(r'\s+', ' ', regex=True)  # Replace multiple spaces with single space
            .str.strip()  # Remove leading/trailing spaces
            .str.replace(' ', '_')  # Replace spaces with underscores
            .str.replace(r'_+', '_', regex=True)  # Remove consecutive underscores
        )
        
    return df_clean
    
cleaned_df = clean_categorical_values(conflict_data, categorical_columns)
```

```python
death_columns = ["one_sided_violence", "non_state", "intrastate", "extrasystemic", "interstate"]
for col in death_columns:
    plt.figure(figsize=(10, 6))
    sns.histplot(conflict_data[col], kde=True, bins=30, color='blue', alpha=0.7)
    plt.title(f"Distribution of {col.replace('_', ' ').title()}")
    plt.xlabel("Death Count")
    plt.ylabel("Frequency")
    plt.grid(True)
    plt.show()
```

```python
# Aggregate death counts by year
yearly_deaths = conflict_data.groupby("year")[death_columns].sum().reset_index()

# Plot yearly trends for each type of violence
plt.figure(figsize=(12, 8))
for col in death_columns:
    plt.plot(yearly_deaths["year"], yearly_deaths[col], marker='o', label=col.replace('_', ' ').title())

plt.title("Yearly Trends of Death Counts by Conflict Type")
plt.xlabel("Year")
plt.ylabel("Death Count")
plt.legend()
plt.grid(True)
plt.show()
```

```python
regional_deaths = conflict_data.groupby("entity")[death_columns].sum().reset_index()

# Create an interactive bar plot using Plotly
fig = px.bar(regional_deaths, 
             x="entity", 
             y=death_columns, 
             title="Total Deaths by Conflict Type and Region",
             labels={"value": "Death Count", "Entity": "Region"},
             barmode="group")
fig.show()
```
