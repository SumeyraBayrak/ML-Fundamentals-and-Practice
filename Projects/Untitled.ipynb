IQR (Interquartile Range) ve Medyan arasındaki ilişkileri anlamak, verinin dağılımı ve merkezi eğilimi hakkında önemli bilgiler sağlar. Bu iki ölçüm arasındaki farklılıkları analiz ederek, verinin nasıl dağıldığını daha net bir şekilde anlayabiliriz.

1. IQR = Medyan Durumu
IQR = Medyan durumu, çok nadir bir durumdur ve birkaç farklı anlam taşıyabilir:
Homojen ve simetrik veri dağılımı: Eğer IQR ve medyan aynıysa, bu genellikle verinin simetrik olduğu ve orta noktada yoğunlaştığı anlamına gelir. Verinin %50'si medyan etrafında simetrik bir şekilde yayılır.
Küçük bir dağılım: Hem IQR'nin hem de medyanın benzer olması, verilerin çok dar bir aralıkta yoğunlaştığını, yani verinin çok düşük bir yayılım gösterdiğini gösterebilir.
Örnek Durum:
Bir şehirdeki işsizlik oranları verileri, çoğunlukla ortalama bir değerin etrafında yoğunlaşmış olabilir. Bu durumda IQR ve medyan birbirine yakın olabilir.

2. IQR > Medyan Durumu
IQR > Medyan durumu, verinin dağılımının daha geniş olduğunu ve medyanın verinin merkezini belirlemede etkili olduğunu ama verinin geniş bir aralıkta yayıldığını gösterir.
Yayılma fazla: Bu durumda, verilerin çoğu medyanın etrafında yoğunlaşmaz. Bunun yerine, veri seti daha geniş bir aralığa yayılır. Yani, verinin çeşitliliği fazladır.
Asimetrik dağılım: IQR'nin medyandan büyük olması, verinin asimetrik olduğunu ve verinin uç noktalarında ağırlıklı olarak büyük farklılıklar olduğunu gösterebilir.
Örnek Durum:
Bir şehirdeki işsizlik oranlarının büyük bir kısmı %5 civarında olsa da, bazı bölgelerde %20'ye kadar çıkabiliyor. Bu durumda IQR büyük olabilir, çünkü verinin uç noktalarında büyük farklar var. Medyan ise bu uç noktalardan etkilenmez ve daha merkezi bir değeri gösterir.

3. IQR < Medyan Durumu
IQR < Medyan durumu, verinin daha dar bir aralığa yayılmaya başladığını gösterir ve verinin merkezi eğiliminin (medyan) daha belirgin olduğunu ifade eder. Bu durumda, verinin çoğu medyan etrafında toplanmış olabilir, ve veri daha homojen hale gelir.
Dar bir dağılım: IQR'nin küçük olması, verinin genellikle dar bir aralığa sıkışmış olduğunu gösterir.
Verinin homojenliği: Bu durumda verinin çoğunluğu medyan etrafında yoğunlaşır, yani uç değerler etkili değildir.
Örnek Durum:
Bir şehirdeki işsizlik oranları çoğunlukla %8-9 arasında yoğunlaşmış olabilir. Bu durumda IQR, medyandan küçük olacaktır çünkü veri dar bir aralığa sıkışmıştır ve işsizlik oranları çoğunlukla medyan civarındadır.

Özetle:
IQR = Medyan: Veri simetrik ve homojendir, yayılım küçük.
IQR > Medyan: Veri daha yayılmış ve asimetriktir, uç noktalar belirgin.
IQR < Medyan: Veri dar bir aralığa yayılmıştır, merkezi eğilim belirgin ve daha homojendir.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample sales data
sales_data = {'Month': pd.date_range(start='2025-01-01', periods=6, freq='M'),
              'Salesperson_A': [5000, 7000, 8000, 6000, 9000, 10000],
              'Salesperson_B': [4000, 6000, 7000, 5000, 8000, 9500]}

# Create a DataFrame
df = pd.DataFrame(sales_data)

# Calculate cumulative sales
df['Cumulative_A'] = df['Salesperson_A'].cumsum()
df['Cumulative_B'] = df['Salesperson_B'].cumsum()

# Plotting
plt.plot(df['Month'], df['Cumulative_A'], marker='o', label='Salesperson A')
plt.plot(df['Month'], df['Cumulative_B'], marker='s', label='Salesperson B')
plt.xlabel('Month')
plt.ylabel('Cumulative Sales')
plt.title('Cumulative Sales Performance')
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
